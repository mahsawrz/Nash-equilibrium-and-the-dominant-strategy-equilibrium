```python
# Making the game matrix
def define_game_matrix():
    num_rows = int(input("Enter the number of rows in the game matrix: "))
    num_cols = int(input("Enter the number of columns in the game matrix: "))

    game_matrix = []
    for _ in range(num_rows):
        row = list(map(int, input(f"Enter the payoffs for Player 1, Row {_ + 1} (separated by spaces): ").split()))
        game_matrix.append(row)

    for _ in range(num_rows):
        row = list(map(int, input(f"Enter the payoffs for Player 2, Row {_ + 1} (separated by spaces): ").split()))
        game_matrix.append(row)

    return game_matrix

# Find the Nash equilibrium
def find_nash_equilibria(game_matrix):
    num_rows = len(game_matrix) // 2
    num_cols = len(game_matrix[0])

    nash_equilibria = []
    for i in range(num_rows):
        for j in range(num_cols):
            is_nash = True
            for k in range(num_rows):
                if game_matrix[k][j] > game_matrix[i][j]:
                    is_nash = False
                    break
            if is_nash:
                for k in range(num_cols):
                    if game_matrix[i][k] > game_matrix[i][j]:
                        is_nash = False
                        break
            if is_nash:
                nash_equilibria.append((i, j))
    return nash_equilibria

# Find the dominant strategy equilibrium
def find_dominant_strategy_equilibria(game_matrix):
    num_rows = len(game_matrix) // 2
    num_cols = len(game_matrix[0])

    dominant_strategy_equilibria_p1 = []
    dominant_strategy_equilibria_p2 = []

    # Check dominant strategies for Player 1
    for col in range(num_cols):
        max_payoff = max(game_matrix[row][col] for row in range(num_rows))
        for row in range(num_rows):
            if game_matrix[row][col] == max_payoff:
                is_dominant = all(game_matrix[row][col] >= game_matrix[row][c] for c in range(num_cols) if c != col)
                if is_dominant:
                    dominant_strategy_equilibria_p1.append((row, col))

    # Check dominant strategies for Player 2
    for row in range(num_rows, num_rows * 2):
        max_payoff = max(game_matrix[row][col] for col in range(num_cols))
        for col in range(num_cols):
            if game_matrix[row][col] == max_payoff:
                is_dominant = all(game_matrix[row][col] >= game_matrix[r][col] for r in range(num_rows) if r != row - num_rows)
                if is_dominant:
                    dominant_strategy_equilibria_p2.append((row - num_rows, col))

    # Find the intersection of dominant strategies for both players
    dominant_strategy_intersection = list(set(dominant_strategy_equilibria_p1) & set(dominant_strategy_equilibria_p2))

    return dominant_strategy_intersection

# Main program
game_matrix = define_game_matrix()

print("Game Matrix:")
num_rows = len(game_matrix) // 2
for i in range(num_rows):
    for j in range(len(game_matrix[0])):
        print(f"({game_matrix[i][j]}, {game_matrix[num_rows + i][j]})", end=" ")
    print()

dominant_strategy_intersection = find_dominant_strategy_equilibria(game_matrix)
print("Dominant Strategy Equilibria:", dominant_strategy_intersection)


nash_equilibria = find_nash_equilibria(game_matrix)
print("Nash Equilibria:", nash_equilibria)

```
